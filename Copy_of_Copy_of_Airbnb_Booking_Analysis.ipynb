#### Chart - 18 (Average Avalability by Naighbourhood and Room Type)

````python
# Avg availability by neighborhood and room type
availability = df.groupby(['neighbourhood_group', 'room_type'])['availability_365'].mean().reset_index()

# Pivot the data to prepare for heatmap
availability_pivot = availability.pivot('neighbourhood_group', 'room_type', 'availability_365')

# Create heatmap
sns.heatmap(availability_pivot, annot=True, fmt='.0f', cmap='YlGnBu')
plt.title('Average Availability by Neighborhood and Room Type')
plt.xlabel('Room Type')
plt.ylabel('Neighborhood')
plt.show()
````
